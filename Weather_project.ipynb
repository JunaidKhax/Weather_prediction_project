```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

FILE = r"C:\Users\jk477\OneDrive\Desktop\weather_data.csv"   # your uploaded CSV file

# ---------------------------------------
# 1. LOAD CSV DATA
# ---------------------------------------
df = pd.read_csv(r"C:\Users\jk477\OneDrive\Desktop\weather_data.csv")

# Keep only the last 100 readings
df = df.tail(100)

# Rename columns for simpler handling
df = df.rename(columns={
    "temperature_C": "temp",
    "humidity_%": "humidity"
})

# Convert timestamp to datetime
df["timestamp"] = pd.to_datetime(df["timestamp"])

# ---------------------------------------
# 2. ANALYTICS
# ---------------------------------------

# 5-point moving average
df["temp_ma5"] = df["temp"].rolling(5).mean()

# Daily min/max
df["date"] = df["timestamp"].dt.date
daily_stats = df.groupby("date")["temp"].agg(["min", "max"])

# Condition frequency
condition_freq = df["condition"].value_counts()

print("\n===== ANALYTICS =====")
print("\nDaily Min/Max:")
print(daily_stats)
print("\nCondition Frequency:")
print(condition_freq)


# ---------------------------------------
# 3. VISUALIZATION
# ---------------------------------------

# Temperature & moving average plot
plt.plot(df["timestamp"], df["temp"], label="Temperature")
plt.plot(df["timestamp"], df["temp_ma5"], label="MA(5)", linestyle="dashed")
plt.xlabel("Time")
plt.ylabel("Temperature (C)")
plt.title("Temperature Trend")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Condition frequency bar chart
condition_freq.plot(kind="bar")
plt.title("Weather Condition Frequency")
plt.xlabel("Condition")
plt.ylabel("Count")
plt.show()
 # your uploaded CSV file

# ---------------------------------------
# 1. LOAD CSV DATA
# ---------------------------------------
df = pd.read_csv(r"C:\Users\jk477\OneDrive\Desktop\weather_data.csv")

# Keep only the last 100 readings
df = df.tail(100)

# Rename columns for simpler handling
df = df.rename(columns={
    "temperature_C": "temp",
    "humidity_%": "humidity"
})

# Convert timestamp to datetime
df["timestamp"] = pd.to_datetime(df["timestamp"])

# ---------------------------------------
# 2. ANALYTICS
# ---------------------------------------

# 5-point moving average
df["temp_ma5"] = df["temp"].rolling(5).mean()

# Daily min/max
df["date"] = df["timestamp"].dt.date
daily_stats = df.groupby("date")["temp"].agg(["min", "max"])

# Condition frequency
condition_freq = df["condition"].value_counts()

print("\n===== ANALYTICS =====")
print("\nDaily Min/Max:")
print(daily_stats)
print("\nCondition Frequency:")
print(condition_freq)


# ---------------------------------------
# 3. VISUALIZATION
# ---------------------------------------

# Temperature & moving average plot
plt.plot(df["timestamp"], df["temp"], label="Temperature")
plt.plot(df["timestamp"], df["temp_ma5"], label="MA(5)", linestyle="dashed")
plt.xlabel("Time")
plt.ylabel("Temperature (C)")
plt.title("Temperature Trend")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Condition frequency bar chart
condition_freq.plot(kind="bar")
plt.title("Weather Condition Frequency")
plt.xlabel("Condition")
plt.ylabel("Count")
plt.show()
```
